```python
import crsq_arithmetic as ari
from qiskit.circuit import QuantumCircuit, QuantumRegister

n=4
m=4
qk = QuantumRegister(n, "k")
qd = QuantumRegister(m, "d")
qctrl = QuantumRegister(1, "ctrl")
qtmp = QuantumRegister(n, "tmp")
qc = QuantumCircuit(qd, qtmp, qctrl, qk)

data = [[1], [3], [5], [7], [9]]
ari.vsqrom1(qc, qk ,[qd], qctrl, qtmp, data)
qc.draw(output='mpl', reverse_bits=True, scale=0.6, filename="output/vsqrom1.png")
```

```python
import crsq_arithmetic as ari
from qiskit.circuit import QuantumCircuit, QuantumRegister

n=4
m=4
qk = QuantumRegister(n, "k")
qd = QuantumRegister(m, "d")
qctrl = QuantumRegister(1, "ctrl")
qtmp = QuantumRegister(n, "tmp")
qc = QuantumCircuit(qd, qtmp, qctrl, qk)

data = [[1], [3], [5], [7], [9], [11], [13], [15]]
ari.vsqrom2(qc, qk ,[qd], qctrl, qtmp, data)
qc.draw(output='mpl', reverse_bits=True, scale=0.5, fold=-1, filename="output/vsqrom2.png")
```
