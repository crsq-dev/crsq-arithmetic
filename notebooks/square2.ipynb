```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from crsq_arithmetic.square2 import unsigned_square2

a = QuantumRegister(4, 'a')
b = QuantumRegister(8, 'b')
cr = QuantumRegister(3, 'cr')
help = QuantumRegister(1, 'help')

qc = QuantumCircuit(a,b,cr,help)

unsigned_square2(qc, a, b, cr,help)

qc.draw(scale=0.4)
```

```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from crsq_arithmetic.square2 import signed_square2

n=6
a = QuantumRegister(n, 'a')
d = QuantumRegister(2*n, 'd')
cr = QuantumRegister(n+1, 'cr')
help = QuantumRegister(1, 'help')

qc = QuantumCircuit(a, d, cr, help)

signed_square2(qc, a, d, cr, help)

qc.draw(scale=0.4, fold=100)
```

```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from crsq_arithmetic import signed_square

n=6
a = QuantumRegister(n, 'a')
d = QuantumRegister(2*n, 'd')
cr1 = QuantumRegister(((n+1)//2)*2-1, 'cr1')
cr2 = QuantumRegister((n//2)*2, 'cr2')

qc = QuantumCircuit(a, d, cr1, cr2)

signed_square(qc, a, d, cr1, cr2, use_gates=True)

qc.draw(scale=0.4,fold=100)
```
